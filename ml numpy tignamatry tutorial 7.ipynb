# find Trignoetry sin(), cos() and tan() using numpy TRIGNOMETRY FUNCTION


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.sin(180)
```

```python
np.cos(180)
```

```python
np.tan(60)
```

```python
np.sin(45)
```

```python
sinx=np.arange(0,3*np.pi,0.1)
print(sinx)
```

```python
siny=np.sin(sinx)
```

```python
print(siny)
```

```python
plt.plot(sinx,siny)
plt.xlabel('sinx')
plt.ylabel('siny')
plt.show()
```

```python
cosy=np.cos(sinx)
print(cosy)
```

```python
plt.plot(sinx,cosy)
plt.show()
```

```python
tany=np.tan(sinx)
tany
```

```python
plt.plot(sinx,tany,marker=".")
plt.show()
```
